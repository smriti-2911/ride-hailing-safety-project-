```python
# 🚀 Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap, MarkerCluster
from wordcloud import WordCloud
from scipy.stats import zscore
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest

# 📌 Load Dataset
df = pd.read_csv("/Users/smriti/ride-hailing-safety-project/datasets/raw/ride_safety_dataset.csv")

# ✅ Dataset Overview
print(df.head())
print("Dataset Shape:", df.shape)
print("Column Names:", df.columns)
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())

print("\nData Types of all columns:")
display(df.dtypes)

# 2. Unique Values Analysis
categorical_columns = df.select_dtypes(include=['object']).columns
print("\nUnique values in categorical columns:")
for col in categorical_columns:
    print(f"\n{col}:")
    print(df[col].value_counts().head())

# 3. Summary Statistics by Category
print("\nSummary Statistics by City:")
display(df.groupby('City').describe())

print("\nSummary Statistics by Crime Type:")
display(df.groupby('Crime_Type').describe())

# 🧼 Data Cleaning & Feature Engineering
df.fillna({
    "Crime_Severity": df["Crime_Severity"].median(),
    "Police_Response_Time_Minutes": df["Police_Response_Time_Minutes"].median(),
    "Driver_Rating": df["Driver_Rating"].median(),
    "Ride_Fare_INR": df["Ride_Fare_INR"].median(),
    "Weather_Temperature_C": df["Weather_Temperature_C"].median(),
    "Traffic_Congestion_Level": df["Traffic_Congestion_Level"].mode()[0]
}, inplace=True)

df["Ride_Safety_Score"] = (
    df["Driver_Rating"] * 2 - df["Crime_Severity"] - df["Police_Response_Time_Minutes"] / 5
)

# 🔥 Outlier Detection & Handling (IQR & Z-Score)
Q1 = df["Ride_Fare_INR"].quantile(0.25)
Q3 = df["Ride_Fare_INR"].quantile(0.75)
IQR = Q3 - Q1
df = df[(df["Ride_Fare_INR"] >= (Q1 - 1.5 * IQR)) & (df["Ride_Fare_INR"] <= (Q3 + 1.5 * IQR))]

df["z_score"] = zscore(df["Driver_Rating"])
df = df[df["z_score"].abs() < 3]  # Remove extreme outliers

# 🔍 Univariate Analysis
plt.figure(figsize=(12, 6))
sns.histplot(df["Ride_Fare_INR"], bins=30, kde=True, color="blue")
plt.title("Ride Fare Distribution")
plt.xlabel("Fare (INR)")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Crime_Severity"], color="red")
plt.title("Crime Severity Boxplot")
plt.show()

crime_types = df["Crime_Type"].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(crime_types, labels=crime_types.index, autopct="%1.1f%%", colors=["red", "orange", "blue"])
plt.title("Crime Type Distribution")
plt.show()

# 🔥 Bivariate Analysis
plt.figure(figsize=(10, 6))
df_numeric = df.select_dtypes(include=[np.number])
sns.heatmap(df_numeric.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Traffic_Congestion_Level"], y=df["Ride_Fare_INR"], alpha=0.5)
plt.title("Traffic Congestion vs Ride Fare")
plt.show()

sns.pairplot(df[["Driver_Rating", "Crime_Severity", "Ride_Fare_INR"]])
plt.show()

# 📍 Crime Hotspot Clustering
map_df = df[["Latitude", "Longitude", "Crime_Type"]].dropna()
m = folium.Map(location=[map_df["Latitude"].mean(), map_df["Longitude"].mean()], zoom_start=11)
marker_cluster = MarkerCluster().add_to(m)

for _, row in map_df.iterrows():
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=f"Crime Type: {row['Crime_Type']}",
        icon=folium.Icon(color="red", icon="exclamation-triangle", prefix="fa")
    ).add_to(marker_cluster)

m.save("reports/crime_hotspots_map.html")

# 🚖 Algorithmic Bias Detection
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["City"], y=df["Ride_Fare_INR"], hue=df["Driver_Rating"])
plt.title("Algorithmic Bias in Ride Pricing")
plt.show()

# 🚔 Police Response Time Efficiency
plt.figure(figsize=(10, 6))
sns.barplot(x=df["Crime_Type"], y=df["Police_Response_Time_Minutes"], palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Police Response Time by Crime Type")
plt.show()

# 🔥 AI-Based Fraudulent Fare Detection
iso_forest = IsolationForest(contamination=0.02, random_state=42)
df["Fare_Anomaly"] = iso_forest.fit_predict(df[["Ride_Fare_INR"]])

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df.index, y=df["Ride_Fare_INR"], hue=df["Fare_Anomaly"], palette={1: "blue", -1: "red"})
plt.title("Fraudulent Ride Fare Detection")
plt.show()

# 🧠 AI-Driven Safety Clustering
df_subset = df[["Crime_Severity", "Driver_Rating", "Ride_Safety_Score"]].dropna()
kmeans = KMeans(n_clusters=3, random_state=42).fit(df_subset)
df["Risk_Cluster"] = kmeans.labels_

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Driver_Rating"], y=df["Crime_Severity"], hue=df["Risk_Cluster"], palette="coolwarm")
plt.title("Driver Rating vs Crime Severity (Clustered)")
plt.show()

# 🚖 Safe vs. Unsafe Routes (Based on Ride Safety Score)
safe_routes = df[df["Ride_Safety_Score"] >= 4.0]
unsafe_routes = df[df["Ride_Safety_Score"] < 2.5]

safe_map = folium.Map(location=[df["Latitude"].mean(), df["Longitude"].mean()], zoom_start=11)
for _, row in safe_routes.iterrows():
    folium.CircleMarker(
        location=[row["Latitude"], row["Longitude"]],
        radius=5, color="green", fill=True, fill_opacity=0.7,
        popup=f"Safe Ride: {row['Ride_Safety_Score']}"
    ).add_to(safe_map)
safe_map.save("reports/safe_routes_map.html")

unsafe_map = folium.Map(location=[df["Latitude"].mean(), df["Longitude"].mean()], zoom_start=11)
for _, row in unsafe_routes.iterrows():
    folium.CircleMarker(
        location=[row["Latitude"], row["Longitude"]],
        radius=5, color="red", fill=True, fill_opacity=0.7,
        popup=f"Unsafe Ride: {row['Ride_Safety_Score']}"
    ).add_to(unsafe_map)
unsafe_map.save("reports/unsafe_routes_map.html")

print("✅ EDA Completed Successfully!")
```
